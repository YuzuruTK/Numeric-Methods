```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Colocar o Intervalo entre os Pontos desejados
intervalo = [0,1]

# X inicial
x0 = 0
# Y inicial
y0 = 0
# intervalo entre valores
h = 0.2


def function_x(x, y):
    # Set da Função
    return (2 * x) - y
```

```python
x_values = [x0]
y_values = [y0]


iteration_range = np.sqrt((intervalo[0] - intervalo[1]) ** 2) / h

values = [[h,x0,y0]]

for i in range(0,int(iteration_range)):

    x = intervalo[0] + h*(i+1) 
    y = y_values[i] + h * function_x(x_values[i], y_values[i])
    
    x_values.append(x)
    y_values.append(y)

    values.append([h,x,y])
```

```python
dataframe = pd.DataFrame(data=values, columns=["h","x","y"])

print(dataframe)

sns.scatterplot(data=dataframe, x="x", y="y")
sns.lineplot(data=dataframe, x="x", y="y")
```
